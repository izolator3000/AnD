```python
# Удалим столбцы, в которых меньше 2/3 данных, так как эти данные 
# важны, но по ним нельзя будет сравнивать квартиры
data = data.drop(['ceiling_height', 'is_apartment', 'balcony', 'parks_nearest', 'ponds_nearest'], axis=1)
```

```python
data.info() # Выведем информацию о датасете до очистки от мусора
```

```python
# Удалим все строки с хотя бы одним NaN, потому что каждый из параметров важен для оценки квартиры 
data = data.dropna()
```

```python
data.duplicated().sum() # Дупликатов нет, а неявные дупликаты не выявить, так как могут быть
                        # две очень похожие квартиры, в одном доме и что-то у них может не совпадать
```

```python
data.info() # Посмотрим на полученные результаты
```

```python
# Заменим тип данных, где это возможно
data['last_price'] = data['last_price'].astype(np.int64)
data['floors_total'] = data['floors_total'].astype(np.int64)
data['airports_nearest'] = data['airports_nearest'].astype(np.int64)
data['cityCenters_nearest'] = data['cityCenters_nearest'].astype(np.int64)
data['parks_around3000'] = data['parks_around3000'].astype(np.int64)
data['ponds_around3000'] = data['ponds_around3000'].astype(np.int64)
data['days_exposition'] = data['days_exposition'].astype(np.int64)
```

### Составим три запроса

```python
# В сумме сколько можно заработать, продав все однокомнотные, двухкомнатные, трёх...
pd.pivot_table(data, index='rooms', values='last_price', aggfunc='sum')
```

```python
# В каких по этажности домах мы продаём квартиры?
pd.pivot_table(data, index='floors_total', values='floor', aggfunc='count')
```

```python
# Какое количество продаваемой площади в каждом регионе?
pd.pivot_table(data, index='locality_name', values='living_area', aggfunc='sum')
```
