```python
data.info() # проведём оценку данных
```

```python
data = data.dropna() # очистим 21525-19351=2174 строчек, в которых отсутствуют значения в колонках daysEmployed и total_income
```

```python
data = data.rename(columns={'daysEmployed': 'days_employed'}) # Приведём название столбца к общему виду
```

### Построим точечные диаграммы (матрицы диаграмм рассеяния):

```python
# Точечная диаграмма ежемесячного дохода от возраста клиента
data.plot(x='dob_years', y='total_income', kind='scatter') 
```

Данная диаграмма рассеивания не показывает зависимость между исследуемыми переменными.

```python
# Точечная диаграмма уровня образования от количества детей клиента
data.plot(x='children', y='education_id', kind='scatter')
```

Данная диаграмма рассеивания не показывает зависимость между исследуемыми переменными.

```python
# Точечная диаграмма стажа в днях от количетсва детей у клиента
data.plot(x='children', y='days_employed', kind='scatter')
```

По полученной диаграмме можно сдлелать вывод, что стаж от количества детей не зависит 

### Построим матрицу диаграмм рассеяния

```python
sns.pairplot(data)
```

Из данной матрицы видно, что корреляция в данном датафрейме отсутствует 

### Вычислим коэффициент корреляции Пирсона для оценки связи между переменными

```python
data.corr()
```

Значений близких к единице, находящихся не на главной диагонали, нет

### Построим heatmap (тепловую карту корреляции):

```python
plt.figure(figsize = (15,8))
sns.heatmap(data.corr(), annot = True, fmt='.1g', square=True)
```
